```python
import matplotlib.pyplot as plt
```

```python
import os
import numpy as np
```

```python
data = []
for fn in os.listdir('result/final'):
    if fn.endswith('.txt'):
        if 'edges' in fn:
            continue
        line = open(os.path.join('result/final', fn), 'r').readline()
        data.append([float(tmp) for tmp in line.split(',')[1:-1]])
```

```python
plt.plot(np.array(data[0])/max(data[0]), 'r.-')
plt.plot(np.array(data[1])/max(data[1]), 'b*-')
plt.plot(np.array(data[2])/max(data[2]), 'g^-') 
```
