## Excercise 1: `bible+shakes.nopunc`

```python
# Open input file
ex1_file = open('bible+shakes.nopunc','r')

# Read file into lines
lines = ex1_file.readlines()
```

```python
# Init RDD
rdd = sc.parallelize(lines)
```

```python
from operator import add

def flattener(text):
    text = text.strip()
    return text.split(' ')

data = rdd.flatMap(flattener)
data.collect()
```

```python
data = data.filter(lambda x: x!='')
data.collect()
```

```python
data = data.map(lambda x: (x, 1))
data.collect()
```

```python
data = data.reduceByKey(add)
data.collect()
```

```python
result = data.collect()
sorted(result, key=lambda x: x[1], reverse=True)[0:10]
```

## Excercise 2: `ecoli.fa`

```python
# Open input file
ex2_file = open('ecoli.fa','r')

# Read file into lines
lines = ex2_file.readlines()
del lines[0]
lines
```

```python
# Init RDD
rdd = sc.parallelize(lines)
```

```python
from operator import add

k = 9

def flattener(text):
    text = text.strip()
    return [(text[i:i+k], 1) for i in range(len(text)-k+1)]

data = rdd.flatMap(flattener)
```

```python
data = data.filter(lambda x: x!='')
```

```python
data = data.map(lambda x: (x, 1))
```

```python
data = data.reduceByKey(add)
data.collect()
```

```python
result = data.collect()
sorted(result, key=lambda x: x[1], reverse=True)[0:10]
```
